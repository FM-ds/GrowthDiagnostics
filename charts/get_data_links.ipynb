```python
import os
import json
import pandas as pd
```

```python
topic = "finance"
rows=[]
for path in os.listdir():
    if not path.endswith(".json") or "_" not in path:
        continue

    with open(f"{path}") as f:
        spec = json.load(f)
    title = spec["title"]['text']
    data_url = spec["data"]["url"]
    topic = path.split("_")[0]

    rows.append({
        "title": title,
        "data_url": data_url,
        "topic": topic
    })

df = pd.DataFrame(rows)

    
```

```python
valid_topics = ["macro", "labour", "investment", "finance", "human"]
for topic in valid_topics:
    print("")
    print(f"{topic.title()}")
    temp_df = df[df["topic"] == topic]
    for row in temp_df.iterrows():
        if "github" in row[1]["data_url"]:
            continue
        print(row[1]["title"], row[1]["data_url"])
```
