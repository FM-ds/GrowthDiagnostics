```python
import pandas as pd
import numpy as np
import altair as alt
import eco_style
import json
alt.themes.enable('report')
import geopandas as gpd
```

# Geodata

```python
itl1 = gpd.read_file("data/geo/International_Territorial_Level_1_January_2021_UK_BUC_2022_4731002879923581517.geojson").to_crs(epsg=4326)
itl2 = gpd.read_file("data/geo/International_Territorial_Level_2_January_2021_UK_BUC_V2_2022_-6059817123640952879.geojson").to_crs(epsg=4326)
itl3 = gpd.read_file("data/geo/International_Territorial_Level_3_January_2021_UK_BUC_V3_2022_8835123939158358283.geojson").to_crs(epsg=4326)
```

# GVA Per Head

```python
# GVA Per Head, Current Prices, 2021

df = pd.read_excel("data/raw/regionalgrossvalueaddedbalancedperheadandincomecomponents.xlsx",
                   sheet_name="Table 2", skiprows=1)

option_sets = [
    {
        "level": "ITL1",
        "geo_df": itl1,
        "val_col": "2021",
        "geo_col": "ITL121CD"
    },
    {
        "level": "ITL2",
        "geo_df": itl2,
        "val_col": "2021",
        "geo_col": "ITL221CD"
    },
    {
        "level": "ITL3",
        "geo_df": itl3,
        "val_col": "2021",
        "geo_col": "ITL321CD"
    }
]

for options in option_sets:
    print("GVA Per Head, Current Prices, 2021", options['level'])

    temp_df = df.query(f"ITL == '{options['level']}'")
    temp_df = options['geo_df'].merge(
        temp_df[['ITL code', options['val_col']]], left_on=options["geo_col"], right_on="ITL code", how="left")
    temp_df['value'] = temp_df[options['val_col']].astype(float)

    map = alt.Chart(temp_df).mark_geoshape(
        stroke="rgb(159, 159, 159)",
        strokeWidth=0.5,
    ).encode(
        color=alt.Color('value:Q', scale=alt.Scale(
            type="log",
            domain=[15_000,  70_000],
            scheme='yellowgreenblue'

        ), title=None, legend=alt.Legend(
            values=[15_000, 30_000, 70_000],
            orient="none",
            direction="horizontal",
            gradientLength=200, 
            title=None,
              legendY=0,
              #(datum.value == 17000 ? '<' : datum.value == 120000 ? '>' : '')
              labelExpr=" if(datum.value==15000, '<', '') + if(datum.value==70000, '>', '') +  '' + format(datum.value, ',')", 
              #labelSeparation=0
              )),
    ).project('equalEarth').properties(
        width=600,
        height=725
    )

    map.display()
```
