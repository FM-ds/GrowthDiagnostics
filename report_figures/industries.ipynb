```python
import pandas as pd
import numpy as np
import altair as alt
import eco_style
alt.themes.enable('report')
```

```python
full_df = pd.read_excel("/Users/finn/Documents/GitHub/GrowthDiagnostics/report_figures/data/raw/regionalgrossvalueaddedbalancedbyindustryandallitlregions-7.xlsx", sheet_name="Table 1c", skiprows=1)
```

```python
df = full_df.copy()
df = df[~df['SIC07 code'].astype(str).str.isnumeric()]
exclude = ['A-E', 'G-T']
df = df[~df['SIC07 code'].str.contains('|'.join(exclude))]
# ITL code	Region name	SIC07 code	SIC07 description
df = df[['ITL code', 'Region name', 'SIC07 code', 'SIC07 description', '2022']]
# lose everything in brackets from the SIC07 code code
df['SIC07 code'] = df['SIC07 code'].str.replace(r"\(.*\)", "")

codes = [
    "A (1-3)",
    "B (5-9)",
    "C (10-33)",
    "D (35)",
    "E (36-39)",
    "F (41-43)",
    "G (45-47)",
    "H (49-53)",
    "I (55-56)",
    "J (58-63)",
    "K (64-66)",
    "L (68)",
    "M (69-75)",
    "N (77-82)",
    "O (84)",
    "P (85)",
    "Q (86-88)",
    "R (90-93)",
    "S (94-96)",
    "T (97-98)"
]

df = df[df['SIC07 code'].isin(codes)]

regions = ['North East', 'North West',
       'Yorkshire and The Humber', 'East Midlands', 'West Midlands',
       'East', 'London', 'South East', 'South West', 'Wales', 'Scotland',
       'Northern Ireland'] #['United Kingdom'

df = df[df['Region name'].isin(regions)]
```

```python


chart = alt.Chart(df.query("`SIC07 code` == 'A (1-3)'")).mark_bar().encode(
    y=alt.Y('Region name', title='', sort="-x"),
    x=alt.X('2022', title='GVA (£m)'),
).properties(
    title='Agriculture, Forestry and Fishing (1-3)'
)

chart
```
